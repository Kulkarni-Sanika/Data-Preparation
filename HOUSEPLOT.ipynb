```python
labelencoder = LabelEncoder()
da['mainroad'] = labelencoder.fit_transform(da['mainroad'])
da['guestroom'] = labelencoder.fit_transform(da['guestroom'])
da['basement'] = labelencoder.fit_transform(da['basement'])
da['hotwaterheating'] = labelencoder.fit_transform(da['hotwaterheating'])
da['airconditioning'] = labelencoder.fit_transform(da['airconditioning'])
da['prefarea'] = labelencoder.fit_transform(da['prefarea'])
da['furnishingstatus'] = labelencoder.fit_transform(da['furnishingstatus'])
da
```
